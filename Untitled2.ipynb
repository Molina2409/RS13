```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Configurar el laberinto
def generar_laberinto(tamaño=100, densidad_obstaculos=0.3):
    laberinto = np.zeros((tamaño, tamaño), dtype=int)
    laberinto[np.random.rand(*laberinto.shape) < densidad_obstaculos] = 1  # Generar obstáculos
    laberinto[0, 0] = 0  # Asegurar que el inicio esté libre
    laberinto[-1, -1] = 2  # Colocar la meta
    return laberinto

# Generar el laberinto
laberinto = generar_laberinto(100, 0.3)

# Crear un colormap personalizado
cmap = ListedColormap([
    "yellow",  # 0: Espacios libres
    "black",   # 1: Obstáculos
    "red",     # 2: Meta
    "blue"     # 3: Camino aprendido (agregado más tarde)
])

# Mostrar el laberinto generado
plt.imshow(laberinto, cmap=cmap)
plt.title("Laberinto Generado (Colores: Amarillo y Rojo)")
plt.show()

# Definir acciones
acciones = {0: (-1, 0), 1: (1, 0), 2: (0, -1), 3: (0, 1)}  # Arriba, Abajo, Izquierda, Derecha

# Validar movimiento
def es_valido(laberinto, estado, accion):
    filas, columnas = laberinto.shape
    movimiento = acciones[accion]
    nuevo_estado = (estado[0] + movimiento[0], estado[1] + movimiento[1])
    return (
        0 <= nuevo_estado[0] < filas and
        0 <= nuevo_estado[1] < columnas and
        laberinto[nuevo_estado] != 1
    )

# Implementar Q-Learning
def q_learning(laberinto, episodios=1000, alpha=0.1, gamma=0.9, epsilon=0.1, max_movimientos=1000):
    q_table = np.zeros((laberinto.shape[0], laberinto.shape[1], len(acciones)))
    meta = tuple(np.argwhere(laberinto == 2)[0])  # Coordenadas de la meta

    for episodio in range(episodios):
        estado = (0, 0)  # Punto inicial
        movimientos = 0

        while estado != meta and movimientos < max_movimientos:
            fila, columna = estado

            # Selección de acción (Exploración-Explotación)
            if np.random.rand() < epsilon:
                accion = np.random.choice(list(acciones.keys()))  # Exploración
            else:
                accion = np.argmax(q_table[fila, columna])  # Explotación

            # Validar y realizar el movimiento
            if es_valido(laberinto, estado, accion):
                movimiento = acciones[accion]
                nuevo_estado = (fila + movimiento[0], columna + movimiento[1])
                recompensa = 1 if nuevo_estado == meta else -0.1

                # Actualizar tabla Q
                q_table[fila, columna, accion] += alpha * (
                    recompensa + gamma * np.max(q_table[nuevo_estado]) - q_table[fila, columna, accion]
                )
                estado = nuevo_estado

            movimientos += 1

        # Mensaje de progreso
        if episodio % 100 == 0:
            print(f"Episodio {episodio}/{episodios} completado.")

    return q_table

# Entrenar el modelo
q_table = q_learning(laberinto, episodios=1000, alpha=0.1, gamma=0.9, epsilon=0.1)

# Evaluar el agente
def probar_agente(laberinto, q_table):
    estado = (0, 0)  # Punto inicial
    meta = tuple(np.argwhere(laberinto == 2)[0])
    camino = [estado]

    while estado != meta:
        fila, columna = estado
        accion = np.argmax(q_table[fila, columna])  # Elegir la mejor acción
        movimiento = acciones[accion]
        nuevo_estado = (fila + movimiento[0], columna + movimiento[1])

        # Si el movimiento no es válido, salir del bucle
        if not es_valido(laberinto, estado, accion):
            print("Movimiento inválido detectado, terminando evaluación.")
            break

        estado = nuevo_estado
        camino.append(estado)

    return camino

camino = probar_agente(laberinto, q_table)

# Mostrar gráficamente el recorrido
def mostrar_camino(laberinto, camino):
    laberinto_copia = laberinto.copy()
    for (fila, columna) in camino:
        if laberinto_copia[fila, columna] == 0:  # No sobrescribir inicio/meta
            laberinto_copia[fila, columna] = 3  # Marcar el camino

    plt.imshow(laberinto_copia, cmap=cmap)
    plt.title("Camino Aprendido")
    plt.show()

mostrar_camino(laberinto, camino)
```
